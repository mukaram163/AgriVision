This notebook performs an initial data sanity check:
- Lists dataset classes
- Displays sample images
- Confirms folder organization and data balance

```python
import os
import matplotlib.pyplot as plt
import random
from PIL import Image
```

```python
data_dir = "../data/raw/PlantVillage"

classes = os.listdir(data_dir)
print("Number of classes:", len(classes))
print("Example classes:", classes[:5])
```

```python
plt.figure(figsize=(10,6))
for i in range(6):
    class_name = random.choice(classes)
    class_path = os.path.join(data_dir, class_name)
    img_name = random.choice(os.listdir(class_path))
    img = Image.open(os.path.join(class_path, img_name))
    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.title(class_name)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

- Number of classes: 16
- Image shape: ~256x256 RGB
- Classes visually verified.
